# Ejercicio 3: Torres de Hanoi.

1.  Disea un algoritmo.

* Algoritmo: Torres de Hanoi.

* Entrada: un valor entero positivo **n** (nmero de discos).

* Salida: numero entero **k**: nmero de movimientos y la secuencia de estos.



2. Implementa el algoritmo en Julia.

```julia
function TH(n, a, b, c)
    if n==1
        println("Mover el disco de la estaca ",a," a la estaca ", c)
        return 1
    else
        x=TH(n-1,a,c,b)
        println("Mover el disco de la estaca ",a," a la estaca ",c)
        y=TH(n-1,b,a,c)
        return x+y+1

    end
    
    
end
```

```julia
a = TH(3, 1,2,3)
print("El nmero de pasos es ",a)
```
